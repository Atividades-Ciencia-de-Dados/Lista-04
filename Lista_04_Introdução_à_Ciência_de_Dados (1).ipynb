####Para facilitar a geração dos gráficos vamos gerar algumas colunas para esse objetivo, basicamente o que foi feito na primeira lista.

#####Quantidade de caracteres

```python
df['caracteres'] = df['text_content_anonymous'].str.len().fillna(0).astype(int)
```

#####Quantidade de palavras

```python
#Coluna com quantidade de palavras
df['words'] = df['text_content_anonymous'].str.split().str.len().fillna(0).astype(int)
```

#####Define como Viral se > 1 (Valor idicado na lista 1)

```python
# Conta quantas vezes cada texto aparece no DataFrame
contagem_textos = df['text_content_anonymous'].map(df['text_content_anonymous'].value_counts())

# Se a contagem for > 1, é viral (1), senão 0
df['viral'] = (contagem_textos > 1).astype(int)
```

#####Quantidade de vezes que o texto aparece

```python
# Conta a frequência de cada texto e mapeia para a coluna original
df['sharings'] = df['text_content_anonymous'].map(df['text_content_anonymous'].value_counts())

# Opcional: Preencher NaN com 0
df['sharings'] = df['sharings'].fillna(0).astype(int)
```

#####Classificando entre Positiva, Negativa e Neutra

```python
def classificar_sentimento(score):
    if score < -0.05:
        return -1  # Negativo
    elif score > 0.05:
        return 1   # Positivo
    else:
        return 0   # Neutro

df['sentiment_coluna_score_sentiment'] = df['score_sentiment'].apply(classificar_sentimento)
```

####20. Os 30 unigramas, bigramas e trigramas mais compartilhados (após a remoção de stop words);

```python
import pandas as pd
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
import string

# Baixar stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('portuguese'))

# Função de limpeza
def limpar_texto(texto):
    # Minúsculas, remoção de pontuação
    texto = texto.lower()
    texto = ''.join([c if c not in string.punctuation else ' ' for c in texto])
    tokens = texto.split()
    # Remoção de stopwords
    tokens = [t for t in tokens if t not in stop_words]
    return ' '.join(tokens)

# Aplicar limpeza
textos = df['text_content_anonymous'].dropna().astype(str)
textos_limpos = textos.apply(limpar_texto)
corpus = textos_limpos.tolist()

# Função para extrair n-gramas
def extrair_top_ngrams(corpus, ngram_range=(1,1), top_n=30):
    vectorizer = CountVectorizer(ngram_range=ngram_range)
    X = vectorizer.fit_transform(corpus)
    freqs = X.sum(axis=0).A1
    termos = vectorizer.get_feature_names_out()
    freq_df = pd.DataFrame({'ngrama': termos, 'frequencia': freqs})
    return freq_df.sort_values(by='frequencia', ascending=False).head(top_n)

# Resultados
top_unigramas = extrair_top_ngrams(corpus, ngram_range=(1,1))
top_bigramas = extrair_top_ngrams(corpus, ngram_range=(2,2))
top_trigramas = extrair_top_ngrams(corpus, ngram_range=(3,3))
```

```python
top_unigramas['tipo'] = 'unigrama'
top_bigramas['tipo'] = 'bigrama'
top_trigramas['tipo'] = 'trigrama'

todos = pd.concat([top_unigramas, top_bigramas, top_trigramas])
todos.to_csv('top_ngrams.csv', index=False)
```
